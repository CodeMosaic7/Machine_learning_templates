Importing the required modules.


```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
```

Importing the dataset

```python
dt=pd.read_csv('Position_Salaries.csv')
x=dt.iloc[:,1:-1].values
y=dt.iloc[:,-1].values
```

Training the model on linear regression.

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

Training Polynomial Regression Model on dataset.

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=5)
x_pf=pf.fit_transform(x)
reg_2=LinearRegression()
reg_2.fit(x_pf,y)
```

Linear Regression visualisation

```python
mp.scatter(x,y,color='Green')
mp.plot(x,reg.predict(x),color='blue')
mp.title('Truth vs Lie(linear regression)')
mp.xlabel('Position Level')
mp.ylabel('Salary')
mp.show()
```

Polynomial Linear Regression visualisation


```python
mp.scatter(x,y,color='pink')
mp.plot(x,reg_2.predict(x_pf),color='blue')
mp.title('Truth vs Lie(Polynomial linear regression)')
mp.xlabel('Position Level')
mp.ylabel('Salary')
mp.show()
```

Prediction Using Linear Regression

```python
reg.predict([[8]])
```

Prediction Using Polynomial Linear regression

```python
reg_2.predict(pf.fit_transform([[8]]))
```
