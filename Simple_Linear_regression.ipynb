Importing all the required libraries.


```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Importing ,splitting of dataset into test and training datasetand fitting of Linear regression model to dataset.

```python
dt=pd.read_csv('Salary_Data.csv')
x=dt.iloc[:,:-1].values
y=dt.iloc[:,-1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
rg=LinearRegression()
rg.fit(x_train,y_train)
y_pred=rg.predict(x_test)
```

Visualising training dataset

```python
mp.scatter(x_train,y_train,color='red')
mp.plot(x_train,rg.predict(x_train),color='yellow')
mp.title('Salary v/s Experience')
mp.xlabel("Years of experince")
mp.ylabel("Salary")
mp.show()
```

```python
mp.scatter(x_test,y_test,color='red')
mp.plot(x_train,rg.predict(x_train),color='blue')
mp.title('Salary v/s Experience')
mp.xlabel("Years of experince")
mp.ylabel("Salary")
mp.show()
```
